IMPORT LIBRERIES +CONFIG

```python
%matplotlib inline
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.manifold import TSNE
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
```

```python
sns.set(rc={"figure.figsize": (20, 8)})
PALETTE = sns.color_palette('deep', n_colors=3)
CMAP = ListedColormap(PALETTE.as_hex())
```

```python
# CONFIG RANDOM STATE 
RANDOM_STATE =205478613
```

LOAD IRIS DATASET

```python
dataset = load_iris()
dataset
```

```python
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
target = 'species'
```

```python
iris = pd.DataFrame(
    dataset.data,
    columns=features)
```

```python
iris[target] = dataset.target
iris.head()
```

definif plot

```python
def plot_iris_2d(x, y, title, xlabel="1st eigenvector", ylabel="2nd eigenvector"):
    sns.set_style("darkgrid")
    
    plt.scatter(x, y,
        c=iris['species'],
        CMAP=CMAP,
        s=70)
    
    plt.title(title, fontsize=20, y=1.03)

    plt.xlabel(xlabel, fontsize=16)
    plt.ylabel(ylabel, fontsize=16)
```

```python
plot_iris_2d(
    x = iris['sepal_length'],
    y = iris['sepal_width'],
    title = 'Plotting first two components',

    xlabel = 'Sepal length',
    ylabel = 'Sepal width')
```

```python
# CONFIG PARAMETERS FOR T-SNE
perplxties =list(range(5,50,20)) 
learning_rates = list(range(10,1000,300))
n_iters =list(range(250,1000,100))

learning_rates
```

```python
# creat t_sne for all combinations
for i in perplxties:
    for j in learning_rates:
        for k in n_iters:
            tsne = TSNE(n_components=2, n_iter=k, random_state=RANDOM_STATE,learning_rate=j,perplexity=i)
            points = tsne.fit_transform(iris[features])
            print(f't-sne vizualization for learning rate: {j} and perplexity: {i} and n-iter: {k}')
            print()
            plot_iris_2d(
                x = points[:, 0],
                y = points[:, 1],
                title = 'Iris dataset visualized with t-SNE')
            plt.show()
```
